#### 소리에 적합한 데이터 증강 기법(Data Augmentation)
- White Noise 추가 하여 원본 오디오 생성

```python
import numpy as np
import soundfile as sf

# 오디오 듣기 위해서 사용할 라이브러리
import IPython.display as ipd
# ipd.Audio(y,rate=sr)
```

```python
# 샘플 Wav 파일 로드
file_path = '../99_Studyfile/02_Sample/blues.00001.wav'
data, sr = sf.read(file_path)

# 원본 오디오 실행
ipd.Audio(data, rate=sr)
```

#### 노이즈 생성

```python
# np.random.uniform() => 0과 1 사이의 균등분포 값(랜덤한)
# np.amax() => 최대값

noise_amp = 0.5*np.random.uniform()*np.amax(data)   # 노이즈 크기를 조절하기 위한 스케일 값
noise = noise_amp * np.random.normal(size=data.shape)

print('noise_amp', noise_amp)
print('noise', noise)
```

#### 노이즈 추가

```python
data_noise = data.astype('float32') + noise.astype('float32')
print(data_noise)
```

#### 노이즈가 추가된 음성 데이터 저장

```python
save_path = '../99_Studyfile/03_Output/sample_noise_add.wav'
sf.write(save_path, data_noise, samplerate=sr)
```

#### 노이즈가 추가된 음성 읽기

```python
# 샘플 Wav 파일 로드
noise_wav = '../99_Studyfile/03_Output/sample_noise_add.wav'
noise_data, noise_sr = sf.read(noise_wav)

# 원본 오디오 실행
ipd.Audio(noise_data, rate=noise_sr)
```
