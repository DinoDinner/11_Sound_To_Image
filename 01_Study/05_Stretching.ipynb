#### 소리 데이터에 적합한 데이터 증강 기법 - Stretching

```python
import numpy as np
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
```

```python
# 1. 샘플 Wav 파일 로드
file_path = '../99_Studyfile/02_Sample/blues.00001.wav'
data, sr = sf.read(file_path)
```

```python
# 2. stretching the audrio signal
data_stretch =  librosa.effects.time_stretch(data, rate=0.8)    # 1보다 크면 오디오 신호가 늘어남.
```

```python
# 결과저장
save_path = '../99_Studyfile/03_Output/sample_stretch.wav'
sf.write(save_path, data_stretch, sr)
```

#### 원본 데이터와 늘린 데이터 비교

```python
plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
librosa.display.waveshow(data, sr=sr, alpha=0.8)
plt.title('Streched Audio')

plt.subplot(1,2,2)
librosa.display.waveshow(data_stretch, sr=sr, alpha=0.8)
plt.title('Streched Audio')
plt.show()
```
